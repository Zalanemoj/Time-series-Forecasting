```python
df_Client = pd.DataFrame({'MT_001': df['MT_001']}, index=df.index)
```

```python
df_Client.head()
```

```python
client_id = 'MT_001'
resample_freq = 'H'
```

```python
series=df_Client[client_id].copy()
```

```python
series=series.resample(resample_freq).mean()
```

```python
series.shape
```

```python
data = pd.DataFrame({'consumption': series})
```

```python
data['hour'] = data.index.hour

# Extract day of week (0=Monday, 6=Sunday) from datetime index
data['day_of_week'] = data.index.dayofweek

# Extract day of month (1-31) from datetime index
data['day_of_month'] = data.index.day

# Extract month (1-12) from datetime index
data['month'] = data.index.month

# Extract quarter (1-4) from datetime index
data['quarter'] = data.index.quarter

# Extract year from datetime index
data['year'] = data.index.year

# Create binary feature: 1 if weekend (Sat/Sun), 0 if weekday
data['is_weekend'] = (data['day_of_week'] >= 5).astype(int)

# --- CYCLICAL ENCODING ---
# Encode hour as sine component (captures cyclical nature of 24-hour day)
data['hour_sin'] = np.sin(2 * np.pi * data['hour'] / 24)

# Encode hour as cosine component (works with sine to represent full cycle)
data['hour_cos'] = np.cos(2 * np.pi * data['hour'] / 24)

# Encode month as sine component (captures cyclical nature of 12 months)
data['month_sin'] = np.sin(2 * np.pi * data['month'] / 12)

# Encode month as cosine component (works with sine to represent full cycle)
data['month_cos'] = np.cos(2 * np.pi * data['month'] / 12)

# --- LAG FEATURES ---
# Create lag feature: consumption from 1 hour ago
data['lag_1'] = data['consumption'].shift(1)

# Create lag feature: consumption from 24 hours ago (same hour yesterday)
data['lag_24'] = data['consumption'].shift(24)

# Create lag feature: consumption from 168 hours ago (same hour last week)
data['lag_168'] = data['consumption'].shift(168)

# --- ROLLING STATISTICS ---
# Calculate 24-hour rolling mean (average of last 24 hours)
data['rolling_mean_24'] = data['consumption'].rolling(window=24).mean()

# Calculate 24-hour rolling standard deviation (volatility of last 24 hours)
data['rolling_std_24'] = data['consumption'].rolling(window=24).std()
```
